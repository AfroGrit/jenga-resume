```python


def get_links_to_movies(soup):
    h3s = soup.find_all('div', attrs={'class': "ipc-title--base"})
    movie_links = {}
    
    for el in h3s:
        # Find the h3 and a tags safely
        h3_tag = el.find('h3', attrs={'class': "ipc-title__text"})
        a_tag = el.find('a')
        
        # Ensure both h3_tag and a_tag are not None before adding them to the dictionary
        if h3_tag and a_tag:
            movie_links[h3_tag.get_text(strip=True)] = a_tag['href']
    
    return movie_links


driver = webdriver.Chrome(options=options)

# Navigate to the IMDB films page
films = 'https://www.imdb.com/search/title/?title_type=feature&release_date=2019-01-01,2024-06-30'
driver.get(films)

# Allow page to load
time.sleep(2)

# Scroll to the "50 more" button to make sure it's visible
try:
    load_more_button = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, "//span[contains(text(), '50 more')]"))
    )
    
    # Scroll into view to make it clickable
    driver.execute_script("arguments[0].scrollIntoView(true);", load_more_button)
    time.sleep(1)  # Give time for the scroll action

    # Wait for it to be clickable and click it
    WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//span[contains(text(), '50 more')]"))
    ).click()
    print("Clicked the '50 more' button.")
    
    # Give time for the content to load
    time.sleep(2)

except Exception as e:
    print(f"Error clicking '50 more' button: {e}")

# Parse the dynamically loaded page source with BeautifulSoup
page_source = driver.page_source
imdb_soup = BeautifulSoup(page_source, 'html.parser')

# Extract film links
film_titles = get_links_to_movies(imdb_soup)

# Print the extracted film links
for title, link in film_titles.items():
    print(f"{title}: {link}")

# Close the browser session
driver.quit()
```
